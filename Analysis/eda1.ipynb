```python
# Filetring df2 according to the major crops in Jharkhand

jharkhand_crops = [
    "Rice", "Wheat", "Maize", "Arhar", "Urad", "Moong", "Gram",
    "Mustard", "Niger", "Til", "Sesame", "Groundnut"
]

filtered_df = df2[df2['Crop'].isin(jharkhand_crops)]
print("Filtered dataset size:", filtered_df.shape)
```

```python
selected_columns = [
    'Soil_Type', 'Rainfall_mm', 'Temperature_Celsius', 'Weather_Condition', 'Crop'
]
model_df = filtered_df[selected_columns]
model_df.head()
```

```python
# Defining the mapping
season_map = {
    "Rice": "Kharif",
    "Maize": "Kharif",
    "Arhar": "Kharif",
    "Urad": "Kharif",
    "Moong": "Kharif",
    "Groundnut": "Kharif",
    "Til": "Kharif",
    "Sesame": "Kharif",
    "Niger": "Kharif",
    "Wheat": "Rabi",
    "Gram": "Rabi",
    "Mustard": "Rabi"
}

filtered_df['Season'] = filtered_df['Crop'].map(season_map)
print(filtered_df[['Crop', 'Season']].drop_duplicates())
```

```python
expected_crops = [
    "Rice", "Wheat", "Maize", "Arhar", "Urad", "Moong", "Gram",
    "Mustard", "Niger", "Til", "Sesame", "Groundnut"
]

present_crops = filtered_df['Crop'].unique().tolist()

missing_crops = [crop for crop in expected_crops if crop not in present_crops]
extra_crops = [crop for crop in present_crops if crop not in expected_crops]

print("Present crops:", sorted(set(expected_crops) - set(missing_crops)))
print("Missing crops:", missing_crops)
print("Extra crops in dataset:", extra_crops)
```
