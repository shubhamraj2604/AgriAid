```python
pivot = jharkhand_crops.pivot_table(
    index='Crop_Year', 
    columns='Crop', 
    values='Annual_Rainfall', 
    aggfunc='mean'
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for crop in pivot.columns:
    plt.plot(pivot.index, pivot[crop], label=crop)

plt.title("Trajectory of Annual Rainfall for Different Crops")
plt.xlabel("Crop Year")
plt.ylabel("Average Annual Rainfall")
plt.legend()
plt.show()
```

```python
seasons = jharkhand_crops['Season'].unique()
print(seasons)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 5, figsize=(20, 6), sharey=True) 
fig.suptitle("Top 5 Crops in Each Season (Jharkhand)", fontsize=16, fontweight='bold')

for ax, season in zip(axes, seasons):   # zip is used to combine pairs of ax , season 
    print(ax)
    top_crops = (  
        jharkhand_crops[jharkhand_crops['Season'] == season]['Crop']
        .value_counts()
        .head(5)
    )

    ax.barh(top_crops.index, top_crops.values, color='teal')
    ax.set_title(season)
    ax.set_xlabel("Count")
    ax.invert_yaxis()  # So the top crop appears at the top

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
